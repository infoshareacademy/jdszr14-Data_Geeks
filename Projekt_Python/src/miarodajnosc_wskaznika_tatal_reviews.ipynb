```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
baza1 = pd.read_csv(r"C:\Users\user\OneDrive\Desktop\jdszr14-Data_Geeks\Projekt_Python\game_data_all.csv", index_col=0)

baza1
```

```python
#właściwy kod na obliczenie wieku gry(Michał)
baza1['release'] = pd.to_datetime(baza1['release'])
baza1['rok'] = baza1['release'].dt.year
baza1['game_age'] = 2024 - baza1['rok']
baza1.drop('rok', axis=1)
```

```python
baza1['game_age'].unique()
```

```python
#Usunięcie niepotrzebnych kolumn
columns_to_drop = ['peak_players', 'store_genres', 'detected_technologies']
baza2 = baza1.drop(columns=columns_to_drop)
baza2
```

```python
#Statystyki opisowe dla total_reviews.
baza2['total_reviews'].describe()
```

```python
# Znalezienie niektórych kolumn dla minimalnej i maksymalnej wartości 'total_reviews'.
selected_columns = ['game', 'primary_genre', 'total_reviews', 'game_age']
min_value = baza2.loc[baza2['total_reviews'] == baza1['total_reviews'].min(), selected_columns]
max_value = baza2.loc[baza2['total_reviews'] == baza1['total_reviews'].max(), selected_columns]
print(min_value)
print(max_value)
```

```python
# Znalezienie 10 najczęsciej powtarzających się 'total_reviews'.
most_common = baza1['total_reviews'].value_counts().head(10)
print(most_common)
```

```python
#Grupowanie po wieku gry i obliczanie średniej liczby recenzji w grupie.
srednia_wg_wieku = baza1.groupby('game_age')[['total_reviews']].mean().reset_index()
srednia_wg_wieku
```

```python
plt.figure(figsize=(8, 5))
plt.bar(srednia_wg_wieku['game_age'], srednia_wg_wieku['total_reviews'], color='skyblue')
plt.xlabel('Game Age')
plt.ylabel('Total Review')
plt.title('Średnia liczba recenzji w zależności od wieku gry')
plt.xticks(srednia_wg_wieku['game_age'])
plt.grid(axis='y')
plt.show()
```

```python
#Grupowanie po rodzaju gry i obliczanie średniej liczby recenzji w grupie.
srednia_wg_rodzaju = baza2.groupby('primary_genre')[['total_reviews']].mean().reset_index()
srednia_wg_rodzaju
```

```python
plt.figure(figsize=(14, 5))
plt.bar(srednia_wg_rodzaju['primary_genre'], srednia_wg_rodzaju['total_reviews'], color='skyblue')
plt.xlabel('Type of Game')
plt.ylabel('Total Review')
plt.title('Średnia liczba recenzji w zależności od rodzaju gry')
plt.xticks(srednia_wg_rodzaju['primary_genre'], rotation='vertical')
plt.grid(axis='y')
plt.show()
```

```python
#Posortowanie bazy2 wg wieku gry i w tych grupach wg rodzaju. Obliczenie średniej total_reviews i uszeregowanie rodzajów od największej średniej.
posortowana_baza2 = baza2[baza2['total_reviews'] >10000].sort_values(by=['game_age', 'primary_genre'])
średnia_baza2 = posortowana_baza2.groupby(['game_age', 'primary_genre'])['total_reviews'].mean().reset_index()
posortowana_średnia = średnia_baza2.sort_values(by=['game_age', 'total_reviews'], ascending=[True, False])
posortowana_średnia
```

```python
# W wykresie wielkość kropek jest proporcjonalne do wieku gry.
plt.figure(figsize=(10,8))
for rodzaj in posortowana_średnia['primary_genre'].unique():
    data = posortowana_średnia[posortowana_średnia['primary_genre'] == rodzaj]
    plt.scatter(data['primary_genre'], data['total_reviews'], s=data['game_age']*5, label=f'Rodzaj {rodzaj}', alpha=0.5)
plt.xlabel('Rodzaj')
plt.ylabel('Liczba Recenzji')
plt.title('Wiek gry a liczba recenzji dla poszczególnych rodzajów')
plt.xticks(rotation='vertical')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Kernel Density Estimate
plt.figure(figsize=(15,6))
baza2['total_reviews'].plot(kind='kde')
plt.xlabel('Total Reviews')
plt.ylabel('Gęstość')
plt.title('Oszacowanie gęstości dla Total Reviews')
plt.show()
```

```python
sns.kdeplot(baza2['total_reviews'], bw_adjust=1.5)
plt.xlabel('Total Reviews')
plt.ylabel('Gęstość')
plt.title('Oszacowanie gęstości dla Total Reviews')
plt.show()
```
